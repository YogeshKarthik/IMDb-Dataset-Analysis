```python
#Top 10 Ancient people in the dataset
df.birthYear.dropna()
df.sort_values("birthYear").head(10)
```

```python
#use=df[["birthYear","deathYear","primaryName"]].dropna()

#Top 10 people with longest lifespans in the dataset

df['birthYear'] = df['birthYear'].replace('\\N', float('NaN'))
df['deathYear'] = df['deathYear'].replace('\\N', float('NaN'))
df['birthYear'] = df['birthYear'].astype(float)
df['deathYear'] = df['deathYear'].astype(float)
use=df[["birthYear","deathYear","primaryName"]].dropna()

use["lifespan"]=use.deathYear-use.birthYear

use["lifespan"][(use.lifespan>200)|(use.lifespan<0)]=use.lifespan.median()
use.sort_values("lifespan",ascending=False).head(10)
```

```python
#Top 10 people with most titles(movies/episodes) with the profession 'director' and 'actor'

df['primaryProfession'] = df['primaryProfession'].astype(str)
result=df.loc[lambda df:df['primaryProfession'].isin(['director','actor'])].sort_values(by='birthYear').groupby('primaryName')['knownForTitles'].count().sort_values(ascending=False).head(10)
print(result)
```

```python
filepath="/content/drive/MyDrive/IMDBdatasets/Dataset_files/titlebasics.tsv"
dftb=pd.read_csv(filepath,delimiter='\t')
dftb.head()
```

```python
merged_df=pd.merge(dftr,dftb,on='tconst')
sorted_df = merged_df.sort_values('numVotes', ascending=False)
top_10_titles = sorted_df.head(10)
print(top_10_titles)
```

```python
#Here we have the top 10 most voted movies or series

top_10_titles_names = top_10_titles[['originalTitle','startYear','averageRating']]
print(top_10_titles_names)
```

```python
#Top 10 most voted Series in the dataset

df_series=merged_df[merged_df['titleType']=='tvSeries']
sorted_df_series=df_series.sort_values('numVotes',ascending=False)
top_10_series=sorted_df_series.head(10)
top_10_series_names=top_10_series[['originalTitle','startYear','averageRating']]
print(top_10_series_names)
```
